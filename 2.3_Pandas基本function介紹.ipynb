```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
print("target_names: "+str(iris['target_names']))
y = pd.DataFrame(iris['target'], columns=['target'])
data = pd.concat([x, y], axis=1)
data.head(3)
```

# Data Calculation

```python
data.apply(lambda x: x+1).head(5) # data 裡的所有元素都加 1
# 前者 x 代表 data 裡的所有元素，x+1 是回傳
```

```python
data['sepal length (cm)'].apply(lambda x: x+2).head(5)
```

```python
data['sepal length (cm)'].map(lambda x: x+2).head(5) # 在對 column 做運算上，apply 跟 map 是一樣的功能
# apply 可以對整個 DataFrame 做運算，map 只能對 column 層級做運算
```

# Data Grouping

```python
data.head()
```

```python
data.groupby(by='target').sum()
```

```python
data.groupby(by='target').mean()
```

```python
for i in data.groupby(by='target'):
    print(i)
```

```python
# 依據 target 分類，找出各個 target 中，sepal length (cm) 第二大的資料
data.groupby(by='target').apply(lambda x: x.sort_values(by=['sepal length (cm)'], ascending=False)[1:2])
```

```python
# 驗證以上是否正確
data[data['target']==0]['sepal length (cm)'].sort_values()
```
